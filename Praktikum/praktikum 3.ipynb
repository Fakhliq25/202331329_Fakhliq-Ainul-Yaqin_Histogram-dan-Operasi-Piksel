```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Membaca data gambar

```python
img = cv2.imread("Gambar Abu.jpg")
```

```python
img.shape
```

```python
[baris, kolom] = img.shape[:2]
```

convert BGR to RGB

```python
#202331329_Fakhliq Ainul YAqin
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

Menampilkan Histogram

```python
#2023313329_Fakhliq Ainul Yaqin
fig, axs = plt.subplots(2,2,figsize=(15,5))

#cara1
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])

#cara2
hist = cv2.calcHist([rgb],[0],None,[256],[0,256])
axs[1,0].imshow(rgb)
axs[1,1].plot(hist)

#menampilkan
plt.show()
```

Operasi Pixel Untuk Meningkatkan kecerahan

```python
#202331329_FAkhliq Ainul Yaqin
beta = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] + beta
        citra_cerah[x,y] = gyx

citra_cerah = citra_cerah.astype(np.uint8)

flg, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.rave(),256,[0,256])

plt.show()
```

Operasi Gabungan antara Meningkatkan Kecerahan dan Meregangkan kontras

```python
#202331329_FAkhliq Ainul Yaqin
beta = 10
alpha = 1.2
citra_hasil = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = alpha * rgb[x,y] + beta
        citra_hasil[x,y] = gyx

citra_hasil = citra_hasil.astype(np.uint8)

flg, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_hasil)
axs[1,1].hist(citra_hasil.ravel(),256,[0,256])

plt.show()
```

Membalik Citra

```python
#202331329_Fakhliq Ainul Yaqin
citra_negatif = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = 255 - rgb[x,y]
        citra_negatif [x,y] = gyx

citra_negatif = citra_negatif.astype(np.uint8)

flg, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_negatif)
axs[1,1].hist(citra_negatif.ravel(),256,[0,256])

plt.show()
```
